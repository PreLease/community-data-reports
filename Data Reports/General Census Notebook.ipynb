# API Testing

### How data is retrieved? 

- The retrieval process is similar to the web address. Majority of the work is figuring out the column of interest 


### How data is stored?

- The data is stored in a single row entry, and each entry is a column


### How data is prepared/ manipulated?

- To replicate the census table, it's best to create a custom table and fill

### What is scalability of reporting?

- From a dataset, you can get any column; however it can be very confusing 
- The recommended method is to have a set of variables for different reports
- For each report, a custom table would need to be made
- For tract data, you would need to run each code separately for each tract
