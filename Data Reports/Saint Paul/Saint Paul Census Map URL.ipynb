```python
#Create an easy to use function


def Map_Web(city='St_Paul'):
    if city=='St_Paul':
        a=St_Paul
        b=St_Paul_Zoom
    elif city=='Mpls_East':
        a= Mpls_East
        b= Mpls_Zoom
    else:
        a= Mpls_West
        b= Mpls_Zoom
    return (print(city + ' Median Rent'), 
            print(Host+a +Meat+ b + TableCol_MedianRent), 
            print(city + ' Total INDIVUDALS below poverty line; (households can be selected)'),
            print(Host+a +Meat+ b + TableCol_IndBelowPov),
            print(city + ' Number of Renter Household'),
            print(Host+a +Meat+ b + TableCol_RentHH), 
            print(city + ' Average Renter Household Size'),
            print(Host+a +Meat+ b + TableCol_RentAvgHHSize), 
            print(city + ' New Renters Location'),
            print(Host+a +Meat+ b + TableCol_NewRenter), 
            print(city + ' Median Gross Rent by New Movers'),
            print(Host+a +Meat+ b + TableCol_MedianRentMover), 
            print(city + ' Renter Median Moved in Year'),
            print(Host+a +Meat+ b + TableCol_MedianYear), 
            print(city + ' Renter BedRooms (Specify number on table)'),
            print(Host+a +Meat+ b + TableCol_Bedrooms), 
            print(city + ' Less than 20K Hoeshold income and spend 30% or more on rent'),
            print(Host+a +Meat+ b + TableCol_Inc20),
            print(city + ' More than 20k and less than 50K Hoeshold income and spend 30% or more rent'),
            print(Host+a +Meat+ b + TableCol_Inc50), 
            print(city + ' Rent Elderly population'),
            print(Host+a +Meat+ b + TableCol_EldPop),
            print(city + ' White (Only) Renters'),
            print(Host+a +Meat+ b + TableCol_White),
            print(city + ' Black (Only) Renters'),
            print(Host+a +Meat+ b + TableCol_Black),
            print(city + ' Asian (Only) Renters'),
            print(Host+a +Meat+ b + TableCol_Asian),
            print(city + ' Hispanic (Only) Renters'),
            print(Host+a +Meat+ b + TableCol_Hispanic),
            print(city + ' Speak Other Languages'),
            print(Host+a +Meat+ b + TableCol_OthLang))
```
